```
import os
import pandas as pd

matrix_list = []
date_list = []
df_final = pd.DataFrame()

directory = '/work/lockdown-mobility/data/state_lex_raw'
for filename in os.listdir(directory):
    date = filename[-14:-4]
    date_list.append(date)
    matrix = pd.read_csv(f'{directory}/{filename}')
    matrix['DATE'] = date
    df_final = pd.concat([df_final, matrix])
    matrix_list.append(matrix)

df_final.to_csv('/work/lockdown-mobility/data/state_lex_processed.csv')
df_final
```
