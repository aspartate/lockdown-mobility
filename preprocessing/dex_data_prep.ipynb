```
import pandas as pd
from scipy.signal import savgol_filter

DEX_all = pd.read_csv('/work/lockdown-mobility/data/state_dex_raw.csv')
```

```
DEX_all['dex_7dma'] = DEX_all.dex.rolling(window=7).mean()
DEX_all['dex_savgol'] = savgol_filter(DEX_all.dex, 51, 3)
DEX_all = DEX_all[DEX_all.state != 'DC']

DEX_all.to_csv('/work/lockdown-mobility/data/state_dex_processed.csv')
DEX_all
```
